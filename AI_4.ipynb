```python
# LP-II Practical
# Information Retrieval using NLP (NLTK)

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter

# Download required NLTK data
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

# ---------------- Input Text ----------------
text = "Artificial Intelligence and Machine Learning are important technologies in modern computer science."

print("Original Text:\n", text)

# ---------------- a) Tokenization ----------------
tokens = word_tokenize(text)
print("\nTokens:\n", tokens)

# ---------------- b) Word Frequency ----------------
freq = Counter(tokens)
print("\nWord Frequency:\n", freq)

# ---------------- c) Remove Stopwords ----------------
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in tokens if word.lower() not in stop_words]
print("\nAfter Removing Stopwords:\n", filtered_words)

# ---------------- d) POS Tagging ----------------
pos_tags = nltk.pos_tag(filtered_words)
print("\nPOS Tags:\n", pos_tags)
```
